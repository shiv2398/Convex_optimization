```python
#2d figure
#function in 2d
def plot_2D(f,gm1,gm2,d_rlist,x0_min=-10,x0_max=10,x1_min=-10,x1_max=10):
   
    x = np.linspace(x0_min,x0_max, 256)
    y = np.linspace(x1_min,x1_max, 256) 
    xx, yy = np.meshgrid(x,y)
    z = f(xx, yy)
    fig = plt.figure(figsize=(10,5),dpi=100)
    ax = plt.subplot(111)
    ax.set_xlabel('x0')
    cset = plt.contourf(x, y, z, 25, alpha=0.6, cmap=cmap)
    ax.set_xlim(x0_min - 1, x0_max + 1)
    ax.set_ylabel('x1')
    ax.set_ylim(x1_min - 1, x1_max + 1)
    plt.annotate('Global_minimum',
    ha = 'center', va = 'bottom',
    xytext = (gm1+2, gm2+2),xy = (gm1,gm2),arrowprops = {'facecolor' : 'black'})
    for i in d_rlist:
      r1=np.array([j[0] for j in i])
      r2=np.array([j[1] for j in i])
      ax.plot(r1, r2, color='red',marker="*",markersize=4)
```

```python
def plot_3D(f,gm1,gm2,d_rlist=None):
    def plots_elevation_and_azim( E,A):
        x0_min,x0_max=-15,15
        x1_min,x1_max=-15,15


        x = np.linspace(x0_min,x0_max, 256)
        y = np.linspace(x1_min,x1_max, 256)
        xx, yy = np.meshgrid(x,y)
        z = f(xx, yy)

        # set up the figure
        fig = plt.figure(figsize=(20,10))
        ax = fig.gca(projection='3d')
        ax.set_xlim(x0_min - 1, x0_max + 1)
        ax.set_ylabel('x1')
        ax.set_ylim(x1_min - 1, x1_max + 1)
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.set_zlabel("f(x,y)")
        ax.plot_surface(xx, yy, z, cmap='jet', alpha = 0.7)
        ax.set_zlim(-1, np.max(z))
        for i in d_rlist:
          r1=np.array([j[0] for j in i])
          r2=np.array([j[1] for j in i])
          ax.plot(r1, r2, f(r1, r2), color='red',marker="*",linewidth=2,markersize=10)
        ax.dist=12
        ax.view_init(E,A)
    return interactive(plots_elevation_and_azim,
                          E=(-180,180,5),
                          A=(-180,180,5)
                          )
```

```python
def ackley( x, a=20, b=0.2, c=2*np.pi ):# ValueError if any NaN or Inf
    n = len(x)
    s1 = sum( x**2 )
    s2 = sum( np.cos( c * x ))
    return -a*np.exp( -b*np.sqrt( s1 / n )) - np.exp( s2 / n ) + a + np.exp(1)
```
